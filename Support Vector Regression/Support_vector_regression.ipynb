```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head(10)
```

```python
xAxis = data.iloc[:, 0:1].values
yAxis = data.iloc[:, -1:].values
```

```python
Sc = StandardScaler()
xSAxis = Sc.fit_transform(xAxis)
ySAxis = Sc.fit_transform(yAxis)
```

```python
svr_reg = SVR(kernel="linear")
svr_reg.fit(xSAxis, ySAxis.ravel())
```

```python
plt.scatter(xSAxis, ySAxis, color="Red") 
plt.scatter(xSAxis, svr_reg.predict(xSAxis), color = "blue")
plt.show()
```

```python
svr_reg2 = SVR(kernel="rbf")
svr_reg2.fit(xSAxis, ySAxis.ravel())
```

```python
fig, ax = plt.subplots(figsize = (6,2.7), layout = 'constrained')
ax.scatter(xSAxis, ySAxis, color="Red") 
ax.scatter(xSAxis, svr_reg2.predict(xSAxis), color = "blue")
plt.show()
```

```python
svr_reg3 = SVR(kernel="sigmoid")
svr_reg3.fit(xSAxis, ySAxis.ravel())
```

```python
fig, ax = plt.subplots(figsize = (6,2.7), layout = 'constrained')
ax.scatter(xSAxis, ySAxis, color="Red") 
ax.scatter(xSAxis, svr_reg3.predict(xSAxis), color = "blue")
plt.show()
```

```python
svr_reg4 = SVR(kernel="poly")
svr_reg4.fit(xSAxis, ySAxis.ravel())
```

```python
fig, ax = plt.subplots(figsize = (6,2.7), layout = 'constrained')
ax.scatter(xSAxis, ySAxis, color="Red") 
ax.scatter(xSAxis, svr_reg4.predict(xSAxis), color = "blue")
plt.show()
```
