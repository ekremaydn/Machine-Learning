```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head(5)
```

```python
xAxis = data.iloc[:, 0:1].values
yAxis = data.iloc[:, -1:].values
```

```python
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(xAxis)
lin_reg = LinearRegression()
lin_reg.fit(x_poly, yAxis)
```

```python
fig, ax = plt.subplots(figsize = (6,2.7), layout = 'constrained')
ax.scatter(xAxis, yAxis, color="Red") # Gerçek değerler
ax.scatter(xAxis, lin_reg.predict(x_poly), color = "blue") # tahmin edilen değerler
plt.show()
```
